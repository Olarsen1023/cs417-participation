```python
%%writefile func.py

def find_letter(string, letter):
  if letter in string:
    print(f"{letter} is in {string}")
  else:
    print(f"{letter} is not in {string}")

message: str = "Avie"
letter: chr = "v"

find_letter(message, letter)
```
