```python
import random

print("W/O seed")
for i in range(0,11):
  num = random.randint(0,10)
  print(num)

print("with seed = 28")
for i in range(0,11):
  random.seed(28)
  print(random.randint(0,10))

print("with seed = 49")
for i in range(0,11):
  random.seed(90)
  print(random.randint(0,10))

```
